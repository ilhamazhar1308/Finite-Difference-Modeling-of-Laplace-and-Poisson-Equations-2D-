#**2D Laplace Method (Heat Transfer)**

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of grid points
n = 100

# Initial temperature distribution
t = np.zeros((n, n))

# -----------------------------
# Boundary Conditions
# (0,y)   = 25
# (1,y)   = 50
# (x,0)   = 0
# (x,1)   = 100
# -----------------------------

# Top boundary
for i in range(n):
    t[i, n-1] = 25

# Left boundary
for i in range(n):
    t[0, i] = 50

# Bottom boundary
for i in range(n):
    t[i, 0] = 0

# Right boundary
for i in range(n):
    t[n-1, i] = 100

# -----------------------------
# Laplace Iteration
# -----------------------------
n_iter = 1000

for it in range(n_iter):

    # Plot every 100 iterations
    if it % 100 == 0:
        plt.imshow(t, cmap='jet', aspect='equal')
        plt.colorbar(label='Temperature (Degree)')
        plt.title(f'2D LAPLACE HEAT TRANSFER (Iteration = {it})')
        plt.show()

    # Laplace finite difference calculation
    for i in range(1, n-1):
        for j in range(1, n-1):
            t[i, j] = 0.25 * (
                t[i+1, j] +
                t[i-1, j] +
                t[i, j+1] +
                t[i, j-1]
            )

# Temperature value at the center of the domain
print("Temperature at the center of the grid:", t[n//2, n//2])

# -----------------------------
# Save results to external file
# -----------------------------
file_name = "Laplace.txt"

with open(file_name, "w") as output:
    for i in range(n):
        for j in range(n):
            output.write(f"{t[i, j]:0.2f}\t")
        output.write("\n")

print("Data successfully saved to Laplace.txt")
```

#**2D Poisson Method**


```python
import numpy as np
import matplotlib.pyplot as plt

# Number of grid points and grid spacing
n = 100
h = 1.0

# Initialize temperature field
t = np.zeros((n, n))

# -----------------------------
# Boundary Conditions
# (0,y)   = 25
# (1,y)   = 50
# (x,0)   = 0
# (x,1)   = 100
# -----------------------------

# Top boundary
for i in range(n):
    t[i, n-1] = 25

# Left boundary
for i in range(n):
    t[0, i] = 50

# Bottom boundary
for i in range(n):
    t[i, 0] = 0

# Right boundary
for i in range(n):
    t[n-1, i] = 100

# -----------------------------
# Poisson source parameter
# -----------------------------
source = np.exp(-10)  # constant heat source

# -----------------------------
# Poisson Iteration
# -----------------------------
n_iter = 1000

for it in range(n_iter):

    # Plot every 100 iterations
    if it % 100 == 0:
        plt.imshow(t, cmap='jet', aspect='equal')
        plt.colorbar(label='Temperature (Degree)')
        plt.title(f'2D POISSON HEAT TRANSFER (Iteration = {it})')
        plt.show()

    # Poisson finite difference scheme
    for i in range(1, n-1):
        for j in range(1, n-1):
            t[i, j] = 0.25 * (
                t[i+1, j] +
                t[i-1, j] +
                t[i, j+1] +
                t[i, j-1]
            ) - 0.25 * h**2 * source

# Temperature value at the center of the domain
print("Temperature at the center of the grid:", t[n//2, n//2])

# -----------------------------
# Save results to external file
# -----------------------------
file_name = "Poisson.txt"

with open(file_name, "w") as output:
    for i in range(n):
        for j in range(n):
            output.write(f"{t[i, j]:0.2f}\t")
        output.write("\n")

print("Data successfully saved to Poisson.txt")
```
